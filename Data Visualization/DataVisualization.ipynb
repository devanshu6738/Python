```python
import pandas as pd
import numpy as np

np.random.seed(42)
n=500
data = {
    "CustomerID": range(1, n+1),
    "Age": np.random.randint(18, 66, n),
    "Gender": np.random.choice(["Male", "Female"], n),
    "Annual_Income_k": np.random.randint(100, 1200, n),
    "Spending_Score": np.random.randint(1, 101, n),
    "Membership": np.random.choice(["Silver", "Gold", "Platinum"], n, p=[0.5, 0.3, 0.2]),
    "Purchase_Frequency": np.random.randint(1, 21, n),
    "Preferred_Category": np.random.choice(
        ["Electronics", "Clothing", "Grocery", "Sports", "Beauty"], n
    ),
    "Churn": np.random.choice([0, 1], n, p=[0.8, 0.2])  # 20% churn rate
}
df=pd.DataFrame(data)

print(df.head())
df['Churn'].value_counts()
```

```python
import numpy as np
import pandas as pd
n=500
data = {
    "CustomerID": range(1, n+1),
    "Age": np.random.randint(18, 66, n),
    "Gender": np.random.choice(["Male", "Female"], n),
    "Annual_Income_k": np.random.randint(100, 1200, n),
    "Spending_Score": np.random.randint(1, 101, n),
    "Membership": np.random.choice(["Silver", "Gold", "Platinum"], n, p=[0.5, 0.3, 0.2]),
    "Purchase_Frequency": np.random.randint(1, 21, n),
    "Preferred_Category": np.random.choice(
        ["Electronics", "Clothing", "Grocery", "Sports", "Beauty"], n
    ),
    "Churn": np.random.choice([0, 1], n, p=[0.8, 0.2])  # 20% churn rate
}
df2=pd.DataFrame(data)
```

Kde OR Density Plot

```python
df2.head()
df2['Age'].plot.density(title="Density plot by Age",figsize=(10,3))
```

```python
df2[['Age','Spending_Score','Annual_Income_k']].plot.density(figsize=(10,3),logx=True)
df2
```

```python
df2.groupby('Gender')['Spending_Score'].plot.density(figsize=(10,3),legend=True,title="Spending score by specific Gender")
```
